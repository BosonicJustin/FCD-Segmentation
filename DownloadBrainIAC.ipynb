```python
import requests

def download_dropbox_folder(dropbox_url, output_path="weights.zip"):
    print("Downloading from:", dropbox_url)
    response = requests.get(dropbox_url)

    if response.status_code == 200:
        with open(output_path, "wb") as f:
            f.write(response.content)
        print(f"✅ Downloaded to {output_path}")
    else:
        print(f"❌ Failed with status code {response.status_code}")

# Use modified ?dl=1 link
dropbox_url = "https://www.dropbox.com/scl/fo/i51xt63roognvt7vuslbl/AG99uZljziHss5zJz4HiFis?rlkey=9w55le6tslwxlfz6c0viylmjb&dl=1"
download_dropbox_folder(dropbox_url)
```
